```python
import pandas as pd
```

```python
snow_df = pd.read_csv('snow_food.csv')
snow_df
```

```python
rain_df = pd.read_csv('rain_food.csv')
rain_df
```

```python
rainsnow_df = pd.concat([rain_df,snow_df])
```

```python
rainsnow = rainsnow_df.drop_duplicates(['단어'], keep='first')
```

```python
rainsnow['날씨구분'] = 'rainsnow'
```

```python
rainsnow.to_csv('rainsnow.csv',index=False)
```
