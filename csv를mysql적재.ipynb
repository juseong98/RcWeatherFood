```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine

# pymysql 세팅
db = pymysql.connect(user = 'root', host = 'localhost', passwd = 'ezen', port = 3306, db = 'food')
cursor = db.cursor()

# csv파일 불러오기
df = pd.read_csv('snow_food.csv',encoding = 'utf-8-sig')
df.columns = ['foodname' ,'food_category', 'kcal', 'protein', 'word_count', 'weather_menu']

# sqlalchemy를 사용해 원하는 database에 csv파일 저장
engine = create_engine("mysql+pymysql://root:"+"ezen"+"@127.0.0.1:3306/food?charset=utf8", encoding = "utf-8")
conn = engine.connect()
df.to_sql(name = "food", con = engine, if_exists = 'append', index = False)
conn.close()

# 저장 확인
sql = "select * from food limit 5"
pd.read_sql(sql,db)
```
