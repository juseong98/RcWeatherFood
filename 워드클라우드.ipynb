```python
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('normal_food.csv')
df
```

```python
wc = df.set_index('단어').to_dict()['빈도수']
```

```python
wordCloud = WordCloud(font_path='malgun',width=2000,height=2000,relative_scaling=0.5,background_color='white').generate_from_frequencies(wc)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(wordCloud)
plt.axis('off')
```
