```python
from ckonlpy.tag import Twitter
from konlpy.tag import Okt
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
from wordcloud import WordCloud
```

```python
Twitter = Twitter()
```

```python
df = pd.read_csv('foodinfo1.csv')
df
df_word = list(df['대표식품명'])
Twitter.add_dictionary(df_word, 'Noun')
```

```python
#블로그 데이터 형태소 분석
an = []
for i in range(0,len(list1)) :
    nouns =  Twitter.nouns(list1[i])
    an.append(nouns)
print(an[2])
```

```python
an
```

```python
#본문 마다 단어와 갯수 구하기 

for i in range(0,len(an)) :
    count = Counter(an[i])
    
    an[i] = count.most_common()
```

```python
an
```

```python
#중복값을 더하기 위해 딕셔너리 형태로 만들어줌 
df = pd.DataFrame()
for i in range(0,len(an)) :
    en = dict(an[i])
    df = df.append({'sentiment' : en}, ignore_index=True)

df
```

```python
#중복 값을 더하는 함수 
from operator import itemgetter
def sum_sentiment(x):
    init_dict = list(x)

    result = {}
    for d in init_dict:
        for k in d.keys():
            result[k] = result.get(k, 0) + d[k]
    return result
```

```python
dic_sum = df.agg({'sentiment' : lambda x : sum_sentiment(x)})
type(dic_sum)
```

```python
#더해진 값이 담긴 데이터프레임을 리스트 형태로 변경
dic_sum = dic_sum.values.tolist()
dic_sum
```

```python
#딕셔너리 형태로 만들어 줌
df = dic_sum[0]
df
```

```python
#단어와 빈도수를 열 이름으로 사용
df = pd.DataFrame(list(df.items()),
                   columns=['단어', '빈도수'])
df
```

```python
#빈도수 내림차순으로 정력
df = df.sort_values(by='빈도수',ascending=False)
df
```
